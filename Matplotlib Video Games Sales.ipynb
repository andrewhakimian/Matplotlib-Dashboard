```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('C:\\Users\\Andrew\\Documents\\MS - Baruch\\Spring 2021\\CIS 9655\\Video_Games_Sales.csv', index_col = 'Release_Year')
df.tail()
```

## Super Mario Bros. is one of the best video games of our generation, there I said it. 
### It is a game that falls in the Genre of platform games. It is a simple video game that consists of two dimensional characters and maps, you must make it to the other side of the map by jumping on platforms, getting upgrades and dodging enemies, without falling to your death or being killed. It was a wildly successul game when it came out, and it encompasses over half of all NES console sales. Platform games are a small genre compared to other genres that exist. However, Super Mario Bros. still managed to have blowout sales when it was released. Through the visuals that I have displayed below you will be able to see the massive success that Super Mario Bros. and that it truly was one of the best games in history.

#### (Two caveats to prevent confusion. First, the "Platform" column refers to the type of console, but there is also a genre of video games called "Platform". Second, The data is missing records of video games from 1988 - 1993 this will affect some of the visuals). 

```python
pltfrmGames = df[df.Genre == "Platform"]
pltfrmGames.head()
```

```python
before00 = df[(df.index <= 2000)]
before00.tail()
```

```python
# The pie chart below gives an understanding for how platform games compare to other genre of games in terms of prevalence.

fig, ax = plt.subplots()

explode = (0, 0, 0, 0, 0, 0, 0.3, 0, 0, 0, 0, 0)
Genres = ["Action", "Sports", "Shooter", "Role-Playing", "Racing", "Misc", "Platform", "Fighting", "Simulation", "Adventure", "Strategy", "Puzzle"]
ax.pie(df["Genre"].value_counts(), labels = Genres, explode=explode, autopct='%1.1f%%')
ax.set_title("Video Game Genres")
```

```python
# The bar chart below displays global sales on each platform from before the year 2000.
# I chose to filter this data from before the year 2000 because it emphasized the NES platform compared to other platforms.
# The NES showed very promising sales and Super Mario Bros. played a key role in its success.

fig, ax = plt.subplots()

before00.groupby(by='Platform')['Global_Sales'].sum().plot.bar(color = "darkred")
ax.set_ylabel("Global Sales (millions)")
ax.set_title("Global Video Game Sales by Platform 1980-2000")
```

```python
# This time series below shows global sales of video games from 1980 to 2016. 
# With ground-breaking sales, it shows what a massive impact Super Mario Bros. truly had in the video game industry.

fig, ax = plt.subplots()

ax.plot(df.index, df["Global_Sales"], color = "slategrey")
ax.set_xlabel("Time (years)")
ax.set_ylabel("Global Sales (millions)")
ax.set_title("Global Video Game Sales 1980-2016")
ax.set_ylim(0,45)
ax.set_xlim(1980, 2016)
ax.annotate('Super Mario Bros',
        xy=(1985,40.24), xytext=(1995, 40),
        arrowprops=dict(facecolor='red', shrink=0.05))
```

```python
# The scatter plot below shows where Super Mario Bros. lands in terms of both user and critic ratings.
# For a platform game from the 80's it found itself to be very popular and well received by the public.

fig, ax = plt.subplots()

ax.scatter(df["Critic_Score"], df["User_Score"], s=10, color = "slategrey")
ax.set_xlabel("Critic Rating")
ax.set_ylabel("User Rating")
ax.set_title("User Ratings vs Critic Ratings")
ax.annotate('Super Mario Bros.', 
            xy=(88, 81), xytext=(35, 85),
            bbox=dict(boxstyle='round, pad=0.2', fc='yellow', alpha=0.8),
            arrowprops=dict(facecolor='darkred', shrink=0.05))
```

```python
# The histogram below shows a spread of the frequency of observations per rating from both critics and users.
# Users tend to be less critical and give higher ratings, while critics are more hesitant to give a higher rating.
# Even will less of a tendency to give higher ratings, critics gave Super Mario Bros. a rating of 88 with users giving an 81.  

fig, ax = plt.subplots()

ax.hist(pltfrmGames["Critic_Score"], bins = 10, label = "Critic Rating", histtype = "step", color="darkred")
ax.hist(pltfrmGames["User_Score"], bins = 10, label = "User Rating", histtype = "step", color="slategrey")
ax.set_xlabel("Rating")
ax.set_ylabel("# of observations")
ax.set_title("Critic & User Ratings for Platform Games")
ax.legend(loc = "upper left")
```

```python
fig, ax = plt.subplots(2, 3, figsize = (15,10))

ax[1,0].scatter(df["Critic_Score"], df["User_Score"], s=10, color = "slategrey")
ax[1,0].set_xlabel("Critic Rating")
ax[1,0].set_ylabel("User Rating")
ax[1,0].set_title("User Ratings vs Critic Ratings")
ax[1,0].annotate('Super Mario Bros.', 
                 xy=(88, 81), xytext=(35, 85),
                bbox=dict(boxstyle='round, pad=0.2', fc='yellow', alpha=0.8),
                arrowprops=dict(facecolor='darkred', shrink=0.05))

before00.groupby(by='Platform')['Global_Sales'].sum().plot.bar(ax=ax[0,1], color = "darkred")
ax[0,1].set_ylabel("Global Sales (millions)")
ax[0,1].set_title("Global Video Game Sales by Platform 1980-2000")

ax[0,2].plot(df.index, df["Global_Sales"], color = "slategrey")
ax[0,2].set_xlabel("Time (years)")
ax[0,2].set_ylabel("Global Sales (millions)")
ax[0,2].set_title("Global Video Games Sales 1980-2016")
ax[0,2].set_ylim(0,45)
ax[0,2].set_xlim(1980, 2016)
ax[0,2].annotate('Super Mario Bros',
            xy=(1985,40.24), xytext=(1995, 40),
            arrowprops=dict(facecolor='red', shrink=0.05))

explode = (0, 0, 0, 0, 0, 0, 0.3, 0, 0, 0, 0, 0)
Genres = ["Action", "Sports", "Shooter", "Role-Playing", "Racing", "Misc", "Platform", "Fighting", "Simulation", "Adventure", "Strategy", "Puzzle"]
ax[0,0].pie(df["Genre"].value_counts(), labels = Genres, explode=explode, autopct='%1.1f%%')
ax[0,0].set_title("Video Game Genres")

ax[1,1].hist(pltfrmGames["Critic_Score"], bins = 10, label = "Critic Rating", histtype = "step", color="darkred")
ax[1,1].hist(pltfrmGames["User_Score"], bins = 10, label = "User Rating", histtype = "step", color="slategrey")
ax[1,1].set_xlabel("Rating")
ax[1,1].set_ylabel("# of observations")
ax[1,1].set_title("Critic & User Ratings for Platform Games")
ax[1,1].legend(loc = "upper left")



plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.4)

fig.delaxes(ax[1][2])

plt.show()
```
